```python
##Fetching Text and Sentiments for applying NLP
def Fetch_Text_sent(data):
    tweet_text = []
    row = twitter_train.shape[0]
    for i in range(row):
        tweet_text.append((twitter_train.loc[i,'text'] , twitter_train.loc[i,'airline_sentiment']))
    return tweet_text
```

```python
##Checking function Fetch_Text_sent()
sample = twitter_train.iloc[0:5,0:]
xtrain = Fetch_Text_sent(sample)
len(xtrain)
```

```python
nlp_xtrain=[('','')]
nlp_xtrain = Fetch_Text_sent(twitter_train.iloc[0:,0:])
nlp_xtrain[0]
```

```python
nlp_xtest=[('','')]
nlp_xtest = Fetch_Text_sent(twitter_test.iloc[0:,0:])
len(nlp_xtest)
```

```python
from nltk import WordNetLemmatizer as WL
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize as WT
from nltk import pos_tag
```

```python
xtrain_word = []
for text in nlp_xtrain:
    xtrain_word.append([WT(text[0]) , text[1]])
```

```python
xtest_word = []
for text in nlp_xtest:
    xtest_word.append([WT(text[0]) , text[1]])
```

```python
xtest_word[0]
```

```python
import string
from nltk.corpus import stopwords
stop = stopwords.words('english')
punc = list(string.punctuation)
stop = stop+punc
```

```python
def get_simple_tag(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_reviews(word):
    output_words = []
    for w in word:
        if w.lower() not in stop:
            pos = pos_tag([w])
            lemm = WL()
            lem_word = lemm.lemmatize(w,pos = get_simple_tag(pos[0][1]))
            output_words.append(lem_word.lower())
    return output_words
```

```python
clean_data = []
for doc in xtrain_word:
    clean_data.append((clean_reviews(doc[0]),doc[1]))
clean_data[0:5]
```

```python
len(clean_data)
```

```python
cleanTest_data = []
for doc in xtrain_word:
    cleanTest_data.append((clean_reviews(doc[0]),doc[1]))
cleanTest_data[0:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
categories = [categ for doc,categ in clean_data] 
```

```python
len(categories)
```

```python
text_docs =  [" ".join(doc) for doc,categ in clean_data] 
```

```python
test_text_docs = [" ".join(doc) for doc,categ in cleanTest_data] 
```

```python
text_docs[0]
```

```python
count_vec = CountVectorizer(stop_words = stop,max_features=3000)
```

```python
x_train = count_vec.fit_transform(text_docs)
```

```python
x_train.shape
```

```python
x_test = count_vec.transform(test_text_docs)
count_vec.get_feature_names()
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
svmclf = svm.SVC()
svmclf.fit(x_train,categories)
```

```python
svmpred = svmclf.predict(x_test)
set(svmpred)
```

```python
import csv


with open('output.csv', 'w') as csvFile:
    writer = csv.writer(csvFile,delimiter=",")
    for r in svmpred:
      writer.writerow([r])
```
